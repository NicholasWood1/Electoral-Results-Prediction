```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
election_data_path = '../../datasets/Clean Datasets/Electoral Data/'
census_data_path = '../../datasets/Clean Datasets/Census Data/'
other_data_path = '../../datasets/Raw Datasets/Other data/'
```

```python
import os
import re

census_df = pd.DataFrame()
census_df_norm = pd.DataFrame()

exclude = ['CED to Age (Individual years).csv']

for index, path in enumerate(os.scandir(census_data_path)):
    if path.name not in exclude:
        c_file = pd.read_csv(path.path, index_col=0)
        cens_code = re.match(r'^[0-9a-zA-z]*?(?= )', c_file.index.name).group(0)
        c_file = c_file.add_prefix(cens_code + ' - ')
        c_file.index.name = 'LGA'
        census_df = census_df.join(c_file, how='outer')
        c_file_norm = c_file.apply(lambda x: x / x.sum(), axis=1)
        census_df_norm = census_df_norm.join(c_file_norm, how='outer', rsuffix='.' + str(index))

census_df.head()
```

```python
v_file = pd.read_csv(election_data_path + 'Final_Vote_Count.csv')
v_file
```

```python
elected = v_file[v_file['Elected'] == 'Y'][['DivisionNm', 'PreferencePercent', 'PartyNm']]
elected.head()
```

```python
elected.set_index('DivisionNm', inplace=True)
```

```python
all_df = census_df.join(elected)
all_df_norm = census_df_norm.join(elected)
```

```python
# Export to csv
all_df.to_csv('../../datasets/Clean Datasets/' + 'CED to All Demographics and Results.csv')
all_df_norm.to_csv('../../datasets/Clean Datasets/' + 'CED to All Demographics and Results Norm.csv')
```
