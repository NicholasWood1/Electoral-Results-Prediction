```python
import numpy as np
import pandas as pd
```

```python
election_data_path = '../../datasets/Raw Datasets/Electoral Data/'
census_data_path = '../../datasets/Raw Datasets/Census Data/'
other_data_path = '../../datasets/Raw Datasets/Other Data/'
```

```python
preferences = pd.read_csv(election_data_path + 'HouseDopByDivisionDownload-24310 (3).csv', engine='python', skiprows=1)
preferences
```

```python
# Removing irrelevant columns
fin1 = preferences.drop(columns=['StateAb', 'DivisionID','CandidateID','Surname', 'GivenNm'])
fin1
```

```python
# Seeing if we have the correct number of divisions should be 151
print(len(set(fin1['DivisionNm'])))
```

```python
find_final_counts = fin1[['CountNumber', 'DivisionNm']]
find_final_counts
```

```python
counter = find_final_counts.drop_duplicates()
counter
```

```python
# Create a list with the highest vote/count number for each electorate
count_number = counter.groupby(['DivisionNm'])['CountNumber'].count()- 1
count_number
```

```python
# Test to see if we get the highest vote counts
count_number['Tangney']
```

```python
Final_Count_Number_DF = pd.DataFrame(count_number)
Final_Count_Number_DF
```

```python
Loop_Column_DF = Final_Count_Number_DF.reset_index()
```

# Dataframe that has DivisionNm to highest vote count

```python
Loop_Column_DF
```

```python
# Test on selecting CountNumber with DivisionNm
fin1[(fin1.DivisionNm == 'Adelaide') & (fin1.CountNumber == 4)]
```

```python
# Test to see if this kind of for loop works as intended, i.e. DivisionNm to highest vote count.
for x in Loop_Column_DF['DivisionNm']:
    print(x, count_number[x])
```

## Looping and creating a list of dataframes that correspond to the final vote counts for each electorate, then concatenating that list into a DataFrame

```python
final_pref_list = []
for x in Loop_Column_DF['DivisionNm']:
    Electorate = fin1[(fin1.DivisionNm == str(x)) & (fin1.CountNumber == count_number[x])]
    final_pref_list.append(Electorate)

The_final_df = pd.concat(final_pref_list)
The_final_df
```

```python
# Irrelevant row removal
The_final_df = The_final_df[The_final_df['CalculationType'] != 'Transfer Count']
```

```python
# Irrelevant row removal
The_final_df = The_final_df[The_final_df['CalculationType'] != 'Transfer Percent']
The_final_df
```

```python
# Extracting only preference percent values
final_vote_percent = The_final_df.iloc[1::2]
final_vote_percent
```

```python
final_vote_percent.reset_index(drop=True)
```

```python
df_percent = final_vote_percent.rename(columns={"CalculationValue" : "PreferencePercent"})
df_percent
```

```python
# Extracting Preference count
clean_final_vote_count = The_final_df.iloc[::2]
```

```python
clean_final_vote_count = clean_final_vote_count.rename(columns={"CalculationValue" : "PreferenceCount"})
```

```python
clean_final_vote_count = clean_final_vote_count.drop(['CalculationType'], axis=1)
clean_final_vote_count
```

```python
clean_final_vote_count.reset_index(drop=True)
```

```python
merged_df = clean_final_vote_count.merge(df_percent)
merged_df
```

```python
# Merging both DataFrames to have PreferenceCount & PreferencePercent as columns
final_merged = merged_df.drop('CalculationType', axis=1)
final_merged
```

```python
#Removing zero's, which leaves us with only the 2-party prefered results for each electorate
final_merged_no_zeros = final_merged[final_merged['PreferencePercent'] != 0]
final_merged_no_zeros 
```

```python
final_merged_no_zeros.reset_index()
```

```python
final_merged_no_zeros.to_csv('../../datasets/Clean Datasets/Electoral Data/Final_Vote_Count.csv')
```
